```python
import numpy as np # import numpy library
import pandas as pd # import pandas library
import seaborn as sns # import seaborn (visualization) library
import matplotlib.pyplot as plt # import matplotlib (visualization) library
```

```python
diamonds = sns.load_dataset('diamonds') # load seaborn dataset named 'diamonds'
diamonds.head() # load the first few lines of the 'diamonds' dataset
```

```python
np.cov(diamonds['carat'], diamonds['price']) # calculate the covariance, the diagonal of the matrix is the covariance of each variable with itself and the other values 
```

```python
diamonds.cov() # covariance between the carat size and price, covariance has no limited range
```

```python
np.corrcoef(diamonds['carat'], diamonds['price']) # calculate correlation, the diagonal of the matrix is the correlation of its variable with itself and the total correlation between carat size and price, correlation has values between minus one and one
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds) # plot the carat size and price on a scatter plot
```

```python
diamonds.corr() # show the correlation between all the numeric variables of the dataframe
```

```python
corrmat = diamonds.corr() # store these values in a variable called 'corrmat'
f, ax = plt.subplots(figsize=(10, 9)) # draw a heatmap for visualizing these values

sns.heatmap(corrmat, ax=ax, cmap="YlGnBu", linewidth=0.1) # columns and rows with the variables as labels and the colors representing the correlation between the variables
```

```python
fig, ax = plt.subplots(figsize=(10.5, 7.2))
sns.scatterplot(x='carat', y='price', data=diamonds, color='#F15B2A', ax=ax) # scatter plot as a bigger size and a different color
```

```python
sns.jointplot(x='carat', y='price', data=diamonds, color='#F15B2A') # visualize joint distribution (scatter plot and histograms) between carat size and price
```

```python
sns.jointplot(x='carat', y='price', data=diamonds, color='#F15B2A', kind='reg') # visualize joint distribution (scatter plot and histograms) between carat size and price with regression
```

```python
sns.jointplot(x='carat', y='price', data=diamonds, color='#F15B2A', kind='hex') # visualize joint distribution (scatter plot and histograms) between carat size and price with hex density distribution
```

```python
sns.jointplot(x='carat', y='price', data=diamonds, color='#F15B2A', kind='kde') # visualize joint (density) distribution (scatter plot and histogram) between carat size and price with level curves
```

```python
from mpl_toolkits.mplot3d import Axes3D # use mpl_toolkits library to plot a 3D diagram
import numpy as np # import numpy library
import pandas as pd # import pandas library
import seaborn as sns # import seaborn (visualization) library
import matplotlib.pyplot as plt # import matplotlib (visualization) library
```

```python
diamonds = sns.load_dataset('diamonds') # load seaborn dataset named 'diamonds'
fig = plt.figure() # set our figure with matplotlib
ax = fig.gca(projection='3d') # define the axis properties, as 3D projection

ax.plot_trisurf(diamonds['carat'], diamonds['depth'], diamonds['price']) # use plot_trisurf function to plot a 3D surface
```

```python
fig = plt.figure(figsize=(15, 4)) # define our plot figure
ax = fig.add_subplot(111, projection='3d') # define the axis properties

ax.bar3d(diamonds['carat'], diamonds['depth'], np.zeros(len(diamonds['price'])), 1, 1, diamonds['price']) # use plotbar3d() function with same data 
```

```python
fig, ax = plt.subplots(figsize=(10.5, 7.2))
sns.boxplot(x='carat', y='depth', hue='price', data=diamonds, ax=ax) # use box plot to visualize data
```
