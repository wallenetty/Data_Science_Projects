```python
# Creating Scatter Plots with Matplotlib
import pandas as pd # import Pandas library
from matplotlib import pyplot as plt # import Matplotlib pyplot library
import numpy as np # import Numpy library
```

```python
data = pd.read_csv('countries.csv') # import our data into Pandas dataframe
data.head() # look at the first few lines of data
```

```python
# look at data for year 1952
data_1952 = data[data.year == 1952] # create a variable to store the data from 1952
data_1952.head() # look at the first few lines of this selected data
```

```python
# create a scatter plot of the relationship of GDP Per Capita and Life Expectancy in 1952
plt.scatter(data_1952.gdpPerCapita, data_1952.lifeExpectancy, 5) # plot GDP, lifeExpectancy, and adjust the size of each dot in the scatter plot
plt.title('GDP Per Capita and Life Expectancy in 1952') # add a title to make the graph easier to understand
plt.xlabel('GDP per Capita ($)') # add a label to the x-axis to make the graph easier to read
plt.ylabel('Life Expectancy') # add a label to the y-axis to make the graph easier to read
plt.show() # show the scatter plot data
```

```python
# finding a correlation between gdpPerCapita and lifeExpectancy in 1952
data_1952.gdpPerCapita.corr(data_1952.lifeExpectancy) # find correlation with the corr() function
```

```python
# look at gdpPerCapita not in a linear scale but in a log scale, we're going to use Numpy (np.log10() function)
np.log10([10, 100, 1000]) # write np.log 10 with the argument being a list of 10, 100, 1000, it will give us 1, 2, and 3
```

```python
# draw a scatter plot not with the raw gdpPerCapita but with the log of gdpPerCapita with the base 10 for the data from 1952
plt.scatter(np.log10(data_1952.gdpPerCapita), data_1952.lifeExpectancy)
plt.show() # show the selected data
```

```python
# find the correlation between the log of gdpPerCapita and lifeExpectancy in 1952
np.log10(data_1952.gdpPerCapita).corr(data_1952.lifeExpectancy) # lof of gdpPerCapita and lifeExpectancy in 1952
```

```python
# compare the log of gdpPerCapita and raw gdpPerCapita and lifeExpectancy, shows that the log of gdpPerCapita is more highly correlated with lifeExpectancy in 1952
data_1952.gdpPerCapita.corr(data_1952.lifeExpectancy) # raw gdpPerCapita and lifeExpectancy in 1952
```

```python
# find the set of all the years that are available in our dataframe
years_sorted = sorted(set(data.year)) # create a variable to store the data collected with: use the set() function to find the set of all the years that are available and make sure that the years are sorted in the correct order with the sorted() function
```

```python
for given_year in years_sorted: # run a for loop for each year
    data_year = data[data.year == given_year] # select the data for this year
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy, 5) # make a scatter plot of gdpPerCapita and lifeExpectancy with plt
    plt.title(given_year) # set the title of this chart
    # set a consistent range for the x-axis and y-axis, and ignore the outlier(s)
    plt.xlim(0, 60000) # set the range for the x-asis to zero to 60000
    plt.ylim(25, 85) # set the range for the y-axis to 25 to 85
    plt.xlabel('GDP Per Capita') # add a label to the x-axis
    plt.ylabel('Life Expectancy') # add a label to the y-axis
    # plt.show() # show the chart
    # export these charts as images, examining how they have changed over time by using plt.savefig instead of with plt.show()
    plt.savefig(str(given_year), dpi=200) # dpi = dots per inch and str() function converts the given_year to a string since savefig() expects a string function
    plt.clf() # we need this line to clear the current plot
```

```python
# Earlier, we ignored the outlier whose gdpPerCapita was over $60000, but you might ask, which country was that one?
data[data.gdpPerCapita > 60000] # find out which country that had a higher gdpPerCapita than $60000
```

```python
# draw a scatter plot for each year with lifeExpectancy and the log of gdpPerCapita
for given_year in years_sorted: # run a for loop for each year
    data_year = data[data.year == given_year] # select the data for this year
    plt.scatter(np.log10(data_year.gdpPerCapita), data_year.lifeExpectancy, 5) # make a scatter plot of gdpPerCapita and lifeExpectancy with plt
    plt.title(given_year) # set the title of this chart
    # set a consistent range for the x-axis and y-axis, to ignore the outlier(s)
    plt.xlim(2, 5) # set the range for the x-asis to 2 to 5
    plt.ylim(25, 85) # set the range for the y-axis to 25 to 85
    plt.xlabel('GDP Per Capita (log scale)') # add a label to the x-axis
    plt.ylabel('Life Expectancy') # add a label to the y-axis
    # plt.show() # show the chart
    # export these charts as images, examining how they have changed over time by using plt.savefig instead of with plt.show()
    plt.savefig('log_' + str(given_year), dpi=200) # dpi = dots per inch and str() function converts the given_year to a string since savefig() expects a string function
    plt.clf() # we need this line to clear the current plot
```

```python
# With the log charts we see the same pattern of continuous improvement that we saw earlier, 
# this time with gdpPerCapita on the log scale instead of a linear scale
# As we saw earlier, you can see that these (log scale) charts show more a linear progression between gdpPerCapita
# and lifeExpectancy. In fact, the log of gdpPerCapita has a higher correlation with lifeExpectancy than the
# raw value of gdpPerCapita every year
```

```python
# Compare the lifeExpectancy of the highest lifeExpectancy countries in 1952
# Select the 10 highest/longest lifeExpectancy countries
top10 = data_1952.sort_values('lifeExpectancy', ascending=False).head(10) # sort the data according to the lifeExpectancy column with the sort_values() function in a descending order, not an ascending order and select the top 10
```

```python
# To make a bar graph of lifeExpectancy for these top 10 longest lifeExpectancy countries in 1952
x = range (10) # we'll need to first call range of 10 and store in the variable called x
plt.bar(x, top10.lifeExpectancy) # create a bar plot
plt.xticks(x, top10.country, rotation='vertical') # add the country labels with xticks (coordinates of labels), labels, and rotation of the labels
plt.title('10 Highest Life Expectancy Countries in 1952') # add a title to chart to make this graph easier to read
plt.ylabel('Life Expectancy in Years') # add a label to the y-axis to make this graph easier to read
plt.show() # show the bar chart
```
