```python
houseAge = data.houseAge # create a shorter variable name to simplify our dataset
```

```python
norm = (houseAge - houseAge.mean()) / houseAge.std() # normalize the data in the houseAge column
norm
```

```python
minmax = (houseAge - houseAge.min()) / (houseAge.max() - houseAge.min()) # normalize all the values in the houseAge column to between 0 and 1
minmax
```

```python
minmax.min() # verify the min value of the minmax dataset
```

```python
minmax.max() # verify the max value of the minmax dataset
```
